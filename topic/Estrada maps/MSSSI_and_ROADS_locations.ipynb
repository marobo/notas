# COMPARE DISTRICT CODE FROM RRMPIS and MSSI DATA

```python
import csv

unique_sucos_roads = set()
unique_district_roads = set()
unique_sucos_mssi = set()
unique_district_mssi = set()

with open('RRMPIS_2014.csv', encoding='utf-8-sig') as fp:
    my_reader = csv.DictReader(fp)
    for row in my_reader:
        unique_sucos_roads.add((row['SUCOCODE'],row['SUCONAME']))
        if row['DISTCODE'] != '0':
            unique_district_roads.add((row['DISTCODE'],row['DISTNAME'].title()))


with open('MSSI_FormChoices.csv', encoding='utf-8-sig') as fp:
    my_reader = csv.DictReader(fp)
    for row in my_reader:
        if row['list_name'] == 'Suco':
            unique_sucos_mssi.add((row['name'],row['label']))
            district_code = row['name'][0] if len(row['name']) == 5 else row['name'][:2]
            district_name = row['municipiu_filter']
            unique_district_mssi.add((district_code, district_name))

       
```
