# HOW MANY SUCO NAME?

```python
all_sucos = []
import csv
with open("RRMPIS_2014.csv") as pf:
    my_reader = csv.DictReader(pf)
    for row in my_reader:
        all_sucos.append(row['SUCONAME'])
my_set = set(all_sucos)
len(my_set)
```

# HOW MANY SUCO CODE?

```python
unique_sucos = set()
import csv
with open("RRMPIS_2014.csv") as pf:
    my_reader = csv.DictReader(pf)
    for row in my_reader:
        unique_sucos.add((row['SUCONAME'], row['SUCOCODE']))
unique_sucos

unique_sucos_mssi= set()
with open("MSSI_FormChoices.csv") as pf:
    my_reader = csv.DictReader(pf)
    for row in my_reader:
        unique_sucos_mssi.add((row['name'], row['label']))

unique_sucos_mssi
```
